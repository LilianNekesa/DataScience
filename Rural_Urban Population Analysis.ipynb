```python
#What was the average female population in Kenya?
Female_County_Average=df.groupby('County')["Female"].mean()
print(Female_County_Average)
```

```python
#What was the average male population in Kenya?
Male_County_Average=df.groupby('County')["Male"].mean()
print(Male_County_Average)
```

```python
#Which 20 districts were the most densely populated Kenya in descending order?
District_populations=df.groupby('District')["Total"].sum()

dense_districts=District_populations.sort_values(ascending=False)
dense_districts.head(20)
```

```python
#Which 30 counties were the most densely populated Kenya in ascending order?
County_populations=df.groupby('County')["Total"].sum()

dense_counties=County_populations.sort_values(ascending=False)
dense_counties.head(30)
```
