```python
#handling the missing data

df_no_na=df.dropna()
df_no_na
```

```python
#fill with a missing value
df=df.fillna(0)
df
```

```python
#Display the top 5 video games by global sale-use nlargest()
df_global = df.nlargest(5, 'Global_Sales')
print('top 5 video games by global sale:',df_global)
```

```python
#Calculate the total sales in North America for the 'Action' genre.
total_sales=df[df['Genre']=='Action']['NA_Sales'].sum()
print('The total sales in North America:',total_sales)
```

```python
#Find the top 3 publishers by global sales
top3_publishers=df.nlargest(3,'Global_Sales')
print('top 3 publishers:', top3_publishers)
```

```python
#Calculate the average sales in Europe for the 'Shooter' genre.
average_sales=df[df['Genre']=='Shooter']['EU_Sales'].mean()
print('The average sales in Europe for the Shooter genre:', average_sales)
```

```python
#Find the game with the highest sales in Japan in the 'Sports' genre

highest_sales=df[df['Genre']=='Sports'].nlargest(1,'JP_Sales')
print(highest_sales)
```
