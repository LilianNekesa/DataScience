```python
import plotly.express as px
import pandas as pd
```

```python
#First I will read the data from the Titanic URL provided.

url="https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
data=pd.read_csv(url)
data.head()
```

```python
data.info()
```

```python
#Create a bar plot to visualize the number of passengers who survived and did not survive, broken down by gender.
fig=px.bar(data, x='Sex', y='Survived', labels="Survived:no of passengers", barmode='group', text='Survived', title='Passengers who survived by gender', color='Survived')
fig.show()
```

```python
#I have used bar graph to visualize the number of passengers who survived in the titanic. From my visualization, it's clear that females survived more than the males. 
```

```python
#Create a histogram to display the distribution of passengers' ages. Use different colors for passengers who survived and did not survive.
fig=px.histogram(data, x='Age', color='Survived', nbins=20,title='Age distribution of passengers')
fig.show()
```

```python
#While comparing the survival rate of passengers in the Titanic by age, the histogram above gives a clear glimpse that the elderly aged between 77 and 82 had only one surviver. Also, there was only one person aged between 72 and 77 but did not survive the tragedy. A total of 9 children aged between 0 and 2.5 years did not survive while only 19 children aged between 2.5 years and 7.5 years survived the tragedy. The number of survivors those aged 12.5years to 17.4 years was 21 teenagers. Additionally, the highest number of non survivors was 81 can be seen at age 17.5 years and 22.4 years whereas the highest number of survivers were 43 aged between 22.5 and 27.4 years.
```

```python
#Create a violin plot to visualize the distribution of fare prices paid by passengers, broken down by their class (1st, 2nd, and 3rd class).
fig=px.violin(data,x='Pclass', y='Fare', title='distribution of fare prices paid by passengers', labels='Pclass',)
fig.show()
```

```python
# From the violin chart above, alot of passengers paid fare for class 3 than class 2 and 1. I have conculuded at that point because class 3 have the largest width and densities  at a fare of 27.9. 1st class had most people saturated at a fare of 43.19.
```

```python
#Generate a scatter plot to explore the relationship between passengers' ages and fare prices, using different colors and symbols for the different classes.
fig=px.scatter(data, x='Age', y='Fare', color='Fare', title='Relationship between passengers ages and fare prices')
fig.show()
```

```python
#By comparing the fare prices of passengers in each group, those aged between 20 and 20 had the highest number of passengers paying fare over 100. Also,  the highest fare was paid for the trip was 512.3292 paid by those aged 35 and 36 respectively.
```

```python
#Use faceting to create a scatter plot matrix displaying the relationship between age, fare, and class for passengers who survived and did not survive.
fig = px.scatter_matrix(data, dimensions=["Age", "Fare", "Pclass"], color='Survived', labels={'Age':'Passengers age', 'Fare':'Fare price', 'Pclass':'passenger class'}, title='Age vs.Fare of passenger class')
fig.show()
```
